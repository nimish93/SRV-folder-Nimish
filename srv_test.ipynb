```python
coo = csc.tocoo(copy=False)
```

```python
 y = pd.DataFrame({'index': coo.row, 'col': coo.col, 'data': coo.data}
                 )[['index', 'col', 'data']].sort_values(['index', 'col']
                 ).reset_index(drop=True)
```

```python
X = df_final[['product', 'product_category', 'income_group']]
y = df_final['Output']
```

```python
X = X.astype("category")
```

```python
X['product_cat'] = X['product'].cat.codes
X['product_category_cat'] = X['product_category'].cat.codes
X['income_cat'] = X['income_group'].cat.codes
```

```python
X["product"].unique()
```

```python
X.head()
```

```python
X.columns
```

```python
X_final = X[['product_cat','product_category_cat', 'income_cat']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_final.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.33, random_state=42)
```

```python

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, bootstrap = True, max_features = 'sqrt')

# Fit on training data

model.fit(X_train,  y_train)
```

```python
rf_predictions = model.predict(X_test)

# Probabilities for each class

#rf_probs = model.predict_proba(test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, rf_predictions)
```
